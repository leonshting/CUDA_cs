```python
import pandas as pd
import sklearn.cluster as KM
import seaborn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('/Users/leonshting/Downloads/crime_data.csv')
```

```python
data
```

```python
seaborn.pairplot(data=data)
```

```python
data.Assault
data.Murder
```

```python
kmeans = KM.KMeans(n_clusters= 10)
```

```python
y = kmeans.fit_predict(data.values)
data['color'] = pd.Series(y)
```

```python
seaborn.pairplot(data, hue= 'color')
```
